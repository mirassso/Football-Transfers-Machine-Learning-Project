```python
import pandas as pd

# Wczytanie danych z pliku CSV
merged_df = pd.read_csv('merged_stats_transfers_2023.csv')
```

```python
# Filtrowanie zawodników, którzy rozegrali co najmniej 400 minut
filtered_df = merged_df[merged_df['Min'] >= 400]
```

```python
# Zapisanie przefiltrowanych danych do nowego pliku CSV
filtered_df.to_csv('filtered_players_400_min.csv', index=False)
```
